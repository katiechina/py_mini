```python
import yfinance as yf

#define the ticker symbol
tickerSymbol = 'AAPL'
# 'MSFT'  'GOOGL'  'AAPL'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
tickerDf = tickerData.history(start='2011-6-22', end='2021-6-22')

tickerDf
```
